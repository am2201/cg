```python
def draw_line(x1, y1, x2, y2, canvas):
    dx = abs(x2 - x1)
    dy = abs(y2 - y1)
    sx = -1 if x1 > x2 else 1
    sy = -1 if y1 > y2 else 1

    if dx > dy:
        err = dx / 2.0
        while x1 != x2:
            canvas[y1][x1] = '*'
            err -= dy
            if err < 0:
                y1 += sy
                err += dx
            x1 += sx
    else:
        err = dy / 2.0
        while y1 != y2:
            canvas[y1][x1] = '*'
            err -= dx
            if err < 0:
                x1 += sx
                err += dy
            y1 += sy
    canvas[y1][x1] = '*'

def scan_fill(poly_points, canvas):
    min_y = min(poly_points, key=lambda p: p[1])[1]
    max_y = max(poly_points, key=lambda p: p[1])[1]

    for y in range(min_y, max_y + 1):
        intersections = []
        for i in range(len(poly_points)):
            x1, y1 = poly_points[i]
            x2, y2 = poly_points[(i + 1) % len(poly_points)]
            if y1 <= y <= y2 or y2 <= y <= y1:
                if y1 != y2:
                    x = int(x1 + (y - y1) * (x2 - x1) / (y2 - y1))
                else:
                    x = x1
                intersections.append(x)
        intersections.sort()
        print("Intersections:", intersections)
        for i in range(0, len(intersections), 2):
            if i + 1 < len(intersections):  # Added check for valid intersection pair
                x1 = intersections[i]
                x2 = intersections[i + 1]
                print("x1:", x1, "x2:", x2)
                for x in range(x1, x2 + 1):
                    canvas[y][x] = '*'


def print_canvas(canvas):
    for row in canvas:
        print(''.join(row))


polygon_points = [(10, 10), (50, 100), (100, 30), (80, 10)]
canvas_width = 150
canvas_height = 150
canvas = [[' ' for _ in range(canvas_width)] for _ in range(canvas_height)]

draw_line(polygon_points[0][0], polygon_points[0][1], polygon_points[1][0], polygon_points[1][1], canvas)
draw_line(polygon_points[1][0], polygon_points[1][1], polygon_points[2][0], polygon_points[2][1], canvas)
draw_line(polygon_points[2][0], polygon_points[2][1], polygon_points[3][0], polygon_points[3][1], canvas)
draw_line(polygon_points[3][0], polygon_points[3][1], polygon_points[0][0], polygon_points[0][1], canvas)

scan_fill(polygon_points, canvas)
print_canvas(canvas)
```
