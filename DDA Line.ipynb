```python
import matplotlib.pyplot as plt
```

```python
x0 = 0
y0 = 0
x1 = 50
y1 = 50
m = (y1-y0)/(x1-x0)
```

```python
x_points = []
y_points = []
while x0<x1:
    x_points.append(x0)
    y_points.append(round(y0))
    x0+=1
    y0+=m
```

```python
plt.scatter(x_points, y_points, c = 'red', s=0.5)
plt.xlim(0,50)
plt.ylim(0,50)
plt.show()
```
